```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
id_ideb = pd.read_csv('inep_ideb.csv', sep = ',')
```

```python
id_ideb
```

```python
redes = id_ideb['rede'].unique()
print(redes)
```

```python
ideb_rio= id_ideb[(id_ideb['ensino']=='fundamental') & (id_ideb['rede']=='municipal') & (id_ideb['ano']>=2010)]

colunas1=['ano', 'rede', 'anos_escolares', 'indicador_rendimento', 'nota_saeb_media_padronizada', 'ideb']

ideb_r= ideb_rio[colunas1]

ideb_r=ideb_r.dropna(axis=0)

ideb_r
```

```python
ideb_r=ideb_r.groupby('ano', as_index=False).mean()
ideb_r
```

```python
ideb_r.plot(x='ano', y='indicador_rendimento', title="Rendimento escolar do Ensino Fundamental Municipal")
```

```python
ideb_rio_est= id_ideb[(id_ideb['ensino']=='fundamental') & (id_ideb['rede']=='estadual') & (id_ideb['ano']>=2010)]

colunas1=['ano', 'rede', 'anos_escolares', 'indicador_rendimento', 'nota_saeb_media_padronizada', 'ideb']

ideb_est= ideb_rio_est[colunas1]

ideb_est=ideb_est.dropna(axis=0)

ideb_est
```

```python
ideb_est=ideb_est.groupby('ano', as_index=False).mean()
ideb_est
```

```python
ideb_est.plot(x='ano', y='indicador_rendimento', title="Rendimento escolar do Ensino Fundamental Estadual")
```

```python
ideb_rio_pr= id_ideb[(id_ideb['ensino']=='fundamental') & (id_ideb['rede']=='federal') & (id_ideb['ano']>=2010)]

colunas1=['ano', 'rede', 'anos_escolares', 'indicador_rendimento', 'nota_saeb_media_padronizada', 'ideb']

ideb_priv= ideb_rio_pr[colunas1]

ideb_priv=ideb_priv.dropna(axis=0)

ideb_priv
```

```python
ideb_priv=ideb_priv.groupby('ano', as_index=False).mean()
ideb_priv
```

```python
ideb_priv.plot(x='ano', y='indicador_rendimento', title="Rendimento escolar do Ensino Fundamental Privado", marker = 'o')
plt.grid(True)
```

```python
plt.plot(ideb_priv['ano'], ideb_priv['indicador_rendimento'], label = 'Privado', marker='x')
```

```python
plt.plot(ideb_est['ano'], ideb_est['indicador_rendimento'], label = 'Estadual', marker='o')
```

```python
plt.plot(ideb_r['ano'], ideb_r['indicador_rendimento'], label = 'Municipal', marker='.')
```

```python
plt.figure(figsize=(10, 6))
ax = plt.gca()

plt.plot(ideb_r['ano'], ideb_r['indicador_rendimento'], label = 'Municipal', marker='x')
plt.plot(ideb_est['ano'], ideb_est['indicador_rendimento'], label = 'Estadual', marker='o')

plt.title('Rendimento escolar Estadual e Municipal')
plt.xlabel('Ano')
plt.ylabel('Rendimento escolar')
plt.legend()
plt.grid(True)  # Adicionar grades
plt.show()
```
