```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
siconfi = pd.read_csv('siconfi.csv', sep = ',')
siconfi_fun = pd.read_csv('siconfi_fun.csv', sep = ',')
siconfi_fun.head()
```

```python
contas = siconfi['conta'].unique()
print(contas)
```

```python
siconfi_desp = siconfi[(siconfi['conta']=='Total Despesa') & (siconfi['ano']==2013) & (siconfi['estagio']=='Despesas Empenhadas')]
siconfi_desp.head(50)
```

```python
siconfi_desp.plot(x='sigla_uf', y='valor', kind = 'bar')
plt.xlabel('Estados')
plt.ylabel('Total empenhado')
plt.title('Despesa total de cada unidade federal')
```

```python
conta = siconfi_fun['conta'].unique()
print(conta)
```

```python
siconfi_edu_BA = siconfi_fun[(siconfi_fun['conta']=='Ensino Fundamental') & (siconfi_fun['estagio_bd']=='Despesas Empenhadas') & (siconfi_fun['sigla_uf']=='BA') & (siconfi_fun['ano']<=2019)]
siconfi_edu_BA.head(50)
```

```python
siconfi_edu_BA = siconfi_edu_BA.sort_values('ano')

plt.plot(siconfi_edu_BA['ano'], siconfi_edu_BA['valor'])

plt.ylabel("milhões (dezena de milhão)", fontsize=10, labelpad=9)

plt.xlabel("ano")

plt.title("Despesas Empenhadas do Ensino Fundamental")
```

```python
siconfi_edu_BA = siconfi_edu_BA.sort_values('ano')

plt.plot(siconfi_edu_BA['ano'], siconfi_edu_BA['valor'])

plt.ylabel("milhões (dezena de milhão)", fontsize=10, labelpad=9)

plt.xlabel("ano")

plt.title("Despesas Empenhadas do Ensino Fundamental")
```
